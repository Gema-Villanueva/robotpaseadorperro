```python
import time

# --- PASO 1: PREPARACIÓN (Checklist) ---
def checklist_robot():
    print("--- INICIANDO CHECKLIST ---")
    gps = True
    microfono = True
    camara = True

    if gps and microfono and camara:
        print("Sistemas OK: Buscando ruta con parque...")
        print("Cálculo de distancia finalizado. Tiempo suficiente.")
        return True
    else:
        print("ERROR: Error en el sistema. Abortando misión.")
        return False

# --- PASO 2: SALIDA DE CASA ---
def preparacion_salida():
    elementos = ["Bolsas de excrementos", "Botella de agua", "Correa"]
    for objeto in elementos:
        print(f"Recogiendo: {objeto}")
    print("Acción: Poniendo correa al perro.")
    print("Acción: Abriendo puerta. ¡En marcha!")

# --- PASO 3: DURANTE EL PASEO (Bucle) ---
def ejecutar_paseo(tiempo_total):
    print("\n--- INICIO DEL PASEO ---")
    cronometro = 0

    while cronometro < tiempo_total:
        print(f"Minuto {cronometro}: Caminando por la ruta planificada...")

        # Simulamos eventos aleatorios durante el paseo
        if cronometro == 2:
            print("[ALERTA] Cámara detecta 'Charco'. Cambiando de acera inmediatamente.")

        if cronometro == 4:
            print("[EVENTO] El perro hace caca. Recogiendo con bolsa y tirando a la papelera.")

        if cronometro == 6:
            print("[EVENTO] El perro hace pis en la pared. Echando agua de la botella.")

        # Control del tiempo para volver
        if cronometro >= tiempo_total - 2:
            print("[SISTEMA] Queda poco tiempo. Girando y volviendo a casa.")
            break

        time.sleep(1) # Simula el paso de un minuto
        cronometro += 1

# --- PASO 4: REGRESO Y LIMPIEZA ---
def finalizar_paseo():
    print("\n--- LLEGADA A CASA ---")
    print("Acción: Quitando correa al perro.")

    patas = [
        "Delantera derecha", "Delantera izquierda",
        "Trasera derecha", "Trasera izquierda"
    ]

    print("Cogiendo toallitas de limpieza...")
    for pata in patas:
        print(f"Limpiando {pata}...")

    print("\n--- FIN DEL ALGORITMO: TAREA COMPLETADA ---")

# --- FLUJO PRINCIPAL DEL ROBOT ---
if checklist_robot():
    preparacion_salida()
    ejecutar_paseo(10) # Paseo de 10 minutos simulados
    finalizar_paseo()
```
